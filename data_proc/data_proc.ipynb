```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('sample.txt', 'r') as f:
    data = f.read()
data = data.replace(' ', '')
data = data.split('\n')[:-1]
data = [d.split(',') for d in data]
data = np.array(data, dtype=int)
data
```

```python
X = data[:, 0][:49]
Y1 = data[:, 1][:49]
Y2 = data[:, 2]
Y3 = data[:, 3][:49]
```

```python
#Y1[:49]
```

```python
plt.figure(figsize=(15, 7))
plt.plot(X, Y1, 'ro', label = 'access latency(time_sender)')
#plt.plot(X, Y2, 'co', label = 'overhead latency(time_STC)')
plt.plot(X, Y3, 'mo', label = 'encoded bits')
plt.axhline(y=80, label = 'threshold')
#plt.axhline(y=1, label = 'y = 1', color='r')
plt.xlabel('number of bits')
plt.ylabel('number of cycles')
plt.legend()
plt.savefig('encoded bits.png')
plt.show()
```
